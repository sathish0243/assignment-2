```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a sample dataset
data = {
    "Category": ["A", "B", "C", "A", "B", "C"],
    "Value": [10, 15, 20, 8, 12, 18]
}

df = pd.DataFrame(data)

# Group by category and create separate histograms
for category, group in df.groupby("Category"):
    plt.hist(group["Value"], label=category, alpha=0.7)

# Add labels and a title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Values Across Categories")
plt.legend()

# Show the plot
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
#heatmap.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)
n_samples = 100
x1 = np.random.normal(size=n_samples)
x2 = x1 + np.random.normal(size=n_samples, scale=0.5)
x3 = np.random.normal(size=n_samples)
x4 = 0.8 * x1 + 0.2 * x2 + np.random.normal(size=n_samples, scale=0.3)

# Create a DataFrame
data = pd.DataFrame({"X1": x1, "X2": x2, "X3": x3, "X4": x4})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="viridis")

# Add a title
plt.title("Correlation Matrix")

# Show the plot
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
#days.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a heatmap matrix
heatmap_data = df.pivot_table(index='Month', columns='Day', values='Sales')

# Create a heatmap
sns.heatmap(heatmap_data, annot=True, cmap='viridis')

# Add labels and a title
plt.xlabel('Day')
plt.ylabel('Month')
plt.title('Sales Variation Heatmap')

# Show the plot
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute z based on the given function
Z = np.sin(np.sqrt(X*2 + Y*2))

# Create a DataFrame for the data
data = {
    'X': X.flatten(),
    'Y': Y.flatten(),
    'Z': Z.flatten()
}
df = pd.DataFrame(data)

# Plot the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.5)

# Add labels and a title
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Population vs. GDP')

# Add country labels as text annotations
for i, row in df.iterrows():
    plt.annotate(row['Country'], (row['GDP'], row['Population']))

# Show the plot
plt.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x
```

```python
from bokeh.plotting import figure, output_file, show
import numpy as np

# Generate x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line to the plot
p.line(x, y, line_width=2, color='blue')

# Output the plot
output_file("sine_wave.html")
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns.
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(10, 100, N)
colors = ["#%06x" % np.random.randint(0, 0xFFFFFF) for _ in range(N)]

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create the plot
p = figure(title="Scatter Plot with Varying Sizes and Colors")

# Add scatter points with the specified sizes and colors
p.circle('x', 'y', source=source, size='sizes', color='colors', alpha=0.5)

# Output the plot
output_file("scatter_plot.html")
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create the plot
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Add a histogram to the plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="black")

# Output the plot
output_file("histogram.html")
show(p)
```
